# RECOMMENDATION THROUGH CORRELATION TECHNIQUE

Welcome to the code notebook for Recommender Systems with Python. This time we will develop basic recommendation systems using Python and pandas.

In this notebook, we will focus on providing a basic recommendation system by suggesting items that are most similar to a particular item, in this case, movies. Keep in mind, this is not a true robust recommendation system, to describe it more accurately,it just tells you what movies/items are most similar to your movie choice.

Let's get started!

## IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
sns.set_style('white')
```

## LOADING DATASET

```python
data = pd.read_csv("u.data")
data.head()
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
data = pd.read_csv("u.data",sep = "\t", names = column_names)
data.head()
```

```python
movie_name = pd.read_csv("Movie_Id_Titles")
movie_name.head()
```

## MERGING BOTH DATASET

```python
data = pd.merge(data,movie_name,on='item_id')
data.head()
```

## RATING 

```python
data.groupby("title")["rating"].mean().sort_values(ascending = False).head()
```

## RATING COUNTS

```python
data.groupby("title")["rating"].count().sort_values(ascending = False).head()
```

```python
rating = pd.DataFrame(data.groupby("title")["rating"].mean())
rating.head()
```

```python
rating["rating_count"] = pd.DataFrame(data.groupby('title')['rating'].count())
rating.head()
```

## DATA VISUALIZATION

```python
plt.figure(figsize=(20,10))
rating["rating_count"].hist(bins = 50)
```

```python
plt.figure(figsize=(20,10))
rating['rating'].hist(bins=50)
```

```python
plt.figure(figsize = (20,10))
sns.jointplot(x='rating',y='rating_count',data=rating,alpha=0.5)
plt.show()
```

```python
data.head()
```

## PIVOT TABLE

```python
movie_pivot = data.pivot_table(index ="user_id", columns= "title",values = "rating")
movie_pivot.head(10)
```

```python
rating.head()
```

```python
rating.sort_values("rating_count", ascending = False).head(10)
```

```python
star_war_rating = movie_pivot["Star Wars (1977)"]
liar_liar_rating = movie_pivot["Liar Liar (1997)"]
star_war_rating.head()
```

```python
star_war_rating.value_counts()
```

```python
similar_to_starwars = movie_pivot.corrwith(star_war_rating)
similar_to_liarliar = movie_pivot.corrwith(liar_liar_rating)
```

```python
similar_to_starwars.sort_values(ascending= False)
```

```python
corr_star_war = pd.DataFrame(similar_to_starwars,columns = ["Correlation"])
corr_star_war.dropna(inplace = True)
corr_star_war.head()
```

```python
corr_star_war.sort_values("Correlation", ascending = False)
```

```python
corr_star_war = corr_star_war.join(rating['rating_count'])
corr_star_war.head(10)
```

```python
corr_star_war[corr_star_war["rating_count"]>100].sort_values('Correlation', ascending = False)
#corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
